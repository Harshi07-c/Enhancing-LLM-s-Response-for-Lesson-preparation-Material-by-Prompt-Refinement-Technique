````python
import pandas as pd

# Function to rank lecture materials for a given query
def rank_lecture_materials(query, zs_material, fs_material, tb_material):
    # Rank lecture materials for the current query
    response = model.generate_content([f""" You are a professor who teaches computer science to students.You need to rank
    lecture preparation materials for a query delimited by triple backticks.The material is generated by 3 different prompt techniques called ZS,FS and TB.Ranking should be on the basis of the content and the structure.
    The material which is the best among  the three should be rank 1 ,followed by the next best material. If you think that 2 materials  similar in terms of content quality and structure, give them the same rank.
    The lecture material are :
    'ZS': {zs_material}
    'FS': {fs_material}
    'TB': {tb_material}
    Now the output should be numerical.
    The ranking should not be similar to the order in which the materials are given to you. Analyze the material on the basis of the criteria give to you and rank them accordingly.
      Query:
          ```{query}```"""])

    # Extract rankings from response
    rankings = response.text.split('\n')

    # Check if response contains rankings
    if len(rankings) == 3:  # Expecting exactly 3 rankings
        # Extracting prompt types from the rankings
        rank1 = rankings[0].split('.')[1].strip()
        rank2 = rankings[1].split('.')[1].strip()
        rank3 = rankings[2].split('.')[1].strip()

        # Add rankings to the final DataFrame
        return {'query': query, '1st': rank1, '2nd': rank2, '3rd': rank3}
    else:
        print("Error: Unable to extract rankings from Gemini response")
        return None

# Create an empty DataFrame to store rankings for all queries
final_df = pd.DataFrame(columns=['query', '1st', '2nd', '3rd'])

# Traverse data and rank lecture materials for each query
for entry in data:
    query = entry['query']
    zs_material = entry['ZS']
    fs_material = entry['FS']
    tb_material = entry['TB']

    # Rank lecture materials for the current query
    query_rankings = rank_lecture_materials(query, zs_material, fs_material, tb_material)

    if query_rankings is not None:
        # Append rankings to the final DataFrame
        final_df = final_df.append(query_rankings, ignore_index=True)

# Display the final DataFrame
print(final_df)
````
