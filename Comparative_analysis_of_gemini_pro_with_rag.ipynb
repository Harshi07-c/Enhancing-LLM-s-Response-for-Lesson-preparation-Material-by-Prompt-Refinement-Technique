```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Count occurrences of FS, TB, and ZS at rank 1
rank_1_counts = df['1st'].value_counts()

# Plot the counts
plt.figure(figsize=(8, 6))
rank_1_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Occurrences of FS, TB, and ZS at Rank 1')
plt.xlabel('Technique')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='')  # Set linestyle to an empty string
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get the counts for each rank and technique
counts = {
    'Rank 1': df['1st'].value_counts(),
    'Rank 2': df['2nd'].value_counts(),
    'Rank 3': df['3rd'].value_counts()
}

# Define the techniques
techniques = ['FS', 'TB', 'ZS']

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Set the width of the bars
bar_width = 0.2

# Set the x locations for the groups
index = np.arange(len(counts))

# Plot each technique
for i, (rank, count) in enumerate(counts.items()):
    ax.bar(index + i * bar_width, count.reindex(techniques, fill_value=0), bar_width, label=rank)

# Customize the plot
ax.set_title('Counts of FS, TB, and ZS Across Ranks')
ax.set_xlabel('Rank')
ax.set_ylabel('Count')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(techniques)
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```

###Topic Combo

```python
topics_combo = [
    "Explain the implementation of stacks using linked lists.",
    "Discuss the applications of arrays in data structures and algorithms.",
    "Illustrate the concept of strings in the context of pattern searching.",
    "Explain how linked lists are utilized in representing polynomial expressions.",
    "Discuss the role of the Dutch National Flag algorithm in sorting algorithms.",
    "Explain how stacks can be implemented iteratively, focusing on the iterative process.",
    "Illustrate the usage of queues in the context of breadth-first search.",
    "Explain the process of hashing and its significance in data storage and retrieval.",
    "Discuss the structure and operations of trees, emphasizing binary trees.",
    "Discuss the concept of search algorithms, including linear and binary search.",
    "Illustrate the working principles of sorting algorithms such as bubble sort and selection sort.",
    "Explain the divide and conquer approach in algorithm design, with examples like merge sort.",
    "Discuss the greedy algorithm paradigm and its applications in optimization problems."
]
# Filter the DataFrame for the comparison queries
tb_df = df[df['query'].isin(topics_combo)]
data = []
# Iterate over each rank
for rank in ['1st', '2nd', '3rd']:
    # Count the occurrences of TB in the specified rank for the filtered topics
    tb_count = tb_df[rank].value_counts().get('TB', 0)
    # Append the data to the list
    data.append({'Rank': rank, 'Count': tb_count})

# Create a DataFrame from the list of dictionaries
tb_count_df = pd.DataFrame(data)

# Plot the count of TB in each rank for the filtered topics
plt.figure(figsize=(8, 6))
sns.barplot(data=tb_count_df, x='Rank', y='Count', palette='viridis')
plt.title('Count of TB in Each Rank for Topics Combination.')
plt.xlabel('Rank')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Initialize counters for TB counts in each rank
tb_count_1st = tb_df['1st'].value_counts().get('TB', 0)
tb_count_2nd = tb_df['2nd'].value_counts().get('TB', 0)
tb_count_3rd = tb_df['3rd'].value_counts().get('TB', 0)

print("TB count for 1st rank:", tb_count_1st)
print("TB count for 2nd rank:", tb_count_2nd)
print("TB count for 3rd rank:", tb_count_3rd)
```
